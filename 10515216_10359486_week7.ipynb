````python
def check_zip(f):
    with zipfile.ZipFile(f) as z:
        for filename in z.namelist():
            return filename
````

````python
def unzip_file(f):
    """Check if file is unzipped if not unzip the file"""
    location = check_zip(f)
    if os.path.exists(location) is True: #check if file is unzipped
        print 'File exist'
    else: #unzip the file if not unzipped
        print 'unzipping file'
        with zipfile.ZipFile(f,'r') as open_zip:
            open_zip.extractall('')
        if os.path.exists(location) is True: #verify file is unzipped
            print 'File has been unzipped'
    !ls -lh Wilders2014/ |head -10
unzip_file('TK_Handelingen_2013-2016.zip')
````
